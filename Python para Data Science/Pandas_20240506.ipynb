```python
#Criando uma matriz 6 x 10, onde 6 é o número de corredores e 10 o número de voltas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

numCorredores = 6
numVoltas = 10
nomesCorredores = ['João', 'José', 'Rayan', 'Lucas', 'Davi', 'Tomás']

velocidade = np.random.uniform(10, size = (numCorredores, numVoltas))
velocidade = np.round(velocidade, 4)

print(velocidade)
```

```python
# Criando um data frame com os dados

tempo = pd.DataFrame(velocidade)
tempo
```

```python
final_pos = pd.DataFrame({'Posicao Final': tempo.index, 'Tempo': posicao.iloc[:, -1]})
final_pos
```

```python
posicao = pd.DataFrame(tempo.cumsum(axis=1))
posicao
```

```python
df = pd.DataFrame(posicao)

# Lista de cores correspondentes
cores = ['blue', 'green', 'red', 'purple', 'orange', 'brown']

# Criar o gráfico de linhas
plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Para cada corredor, plotar os tempos em função do número da volta
for corredor, nome, cor in zip(range(len(df)), nomesCorredores, cores):
    plt.plot(df.columns, df.iloc[corredor], marker='o', label=nome, color=cor)

# Adicionar rótulos e legendas
plt.xlabel('Número da Volta')
plt.ylabel('Tempo (s)')
plt.title('Tempos dos Corredores por Volta')
plt.legend()

# Mostrar o gráfico
plt.grid(True)  # Adiciona uma grade ao gráfico
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series[0]
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 0}, axis=1)
              .sort_values(0, ascending=True))
  xs = counted[0]
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values(0, ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel(0)
_ = plt.ylabel('count()')
```

```python
#Criando uma matriz 6 x 10, onde 6 é o número de corredores e 10 o número de voltas

import numpy as np
import pandas as pd

numCorredores = 6
numVoltas = 10

velocidade = np.random.uniform(100, size = (numCorredores, numVoltas))
velocidade = np.round(velocidade, 4)

print(velocidade)
```

```python
# Criando um data frame com os dados

tempo = pd.DataFrame(velocidade)
tempo
```

```python
# Alterando a descrição das colunas e dos índices

tempo = pd.DataFrame(velocidade, columns = ['Volta 1', 'Volta 2', 'Volta 3', 'Volta 4', 'Volta 5', 'Volta 6', 'Volta 7', 'Volta 8', 'Volta 9', 'Volta 10'], index = ['João', 'José', 'Rayan', 'Lucas', 'Davi', 'Neto'])
tempo
```

```python
# Informar o tempo da volta mais rápida

melhorTempo = tempo.min().min()
melhorTempo
```

```python
# Infomar o índice da volta mais rápida

voltaMaisRapida = tempo.idxmin()
voltaMaisRapida

```

```python
soma = tempo.sum(axis = 1)
soma
```

```python
vencedor = soma.idxmin()
vencedor

```
